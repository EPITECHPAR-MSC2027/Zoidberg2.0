```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Affichre les résultats
def metrics_KNN(y_pred, y_test):
    for i, (pred, real) in enumerate(zip(y_pred, y_test)):
        statut = "✅" if pred == real else "❌"
        print(f"Image {i+1} : Prédit={pred}, Réel={real} {statut}")

    accuracy_with_lda = accuracy_score(y_test, y_pred)
    print(f"\nPrécision finale : {accuracy_with_lda * 100:.2f}%")

# KNN sans réduction
def classic_KNN(X_train_scaled,y_train,X_test_scaled, y_test):

    # Création d'un nouvel objet en définissant les hyperparamétre.
    # Ici on défini juste le nombre maximal de composants
    knn = KNeighborsClassifier(n_neighbors=0.95)

    # Entraîner le model
    knn.fit(X_train_scaled, y_train)

    # Prédire les données
    y_pred = knn.predict(X_test_scaled)

    # Stocker les données pour en faire un diagramme d'accuracy
    accuracy_without_reduction = accuracy_score(y_test, y_pred)
    return accuracy_without_reduction
```

Ce code utilise l'algorithme de régression basé sur le model KNN (K-nearest neighbors) pour s'entraîner sur 4,705 images. Ensuite il fera une prédiction sur les 624 radiographie pulmonaire.

Que fait chaque méthode ?
    - Fit : Ne fait pas vraiment de calcul, il se contente de mémoriser les données d'entraînement.
    - Predict : calcule les distances et trouve les K plus proches voisins
