```python
import os
from dotenv import load_dotenv
from datasets import load_dataset
from huggingface_hub import login

# Charger les variables d'environnements du système.
load_dotenv()
token = os.getenv('HUGGINGFACE_TOKEN')
dataset_name = os.getenv('HUGGINGFACE_DATASET_NAME')    

def setup_huggingface_auth():
    if not token:
        raise ValueError("❌ HUGGINGFACE_TOKEN not found in .env file")
    # Se connecter à HuggingFace lorsque le token a été récupéré avec succès.
    login(token=token)


def load_chest_xray_dataset():
    if not dataset_name:
        raise ValueError("❌ HUGGINGFACE_DATASET_NAME not found in .env file")
    try:
        dataset = load_dataset(dataset_name, token=True)
        return dataset
    
    except Exception as e:
        print(f"❌ Erreur lors du chargement: {e}")
        raise
```

Ce code permet de charger le dataset de notre repo privé sur huggingface.
