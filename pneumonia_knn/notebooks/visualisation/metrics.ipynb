```python
import matplotlib.pyplot as plt

def show_scores(metrics, title):
    labels = [
        f"k={p['knn__n_neighbors']}\n{p['knn__metric']}\npca={p['pca__n_components']}" 
        for p in metrics['params']
    ]

    x = range(len(labels))

    # Graphique
    fig, ax = plt.subplots(figsize=(14, 6))

    ax.plot(x, metrics['mean_test_accuracy'],  marker='o', label='Accuracy')
    ax.plot(x, metrics['mean_test_precision'], marker='s', label='Precision')
    ax.plot(x, metrics['mean_test_recall'],    marker='^', label='Recall')

    ax.set_xticks(x)
    ax.set_xticklabels(labels, fontsize=8)
    ax.set_ylabel('Score')
    ax.set_title(title)
    ax.legend()
    ax.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()
```
